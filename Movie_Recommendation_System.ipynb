```python
# importing libraries

import pandas as pd
import numpy as np
```

```python
# reading files

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies.head() # checking data
```

```python
ratings.head()
```

```python
# preparing data table for analysis

ratings_pivot = pd.pivot_table(data=ratings, values='rating', index='userId', columns='movieId')
ratings_pivot.head()
```

```python
# finding correlation between two variavles. 
# We are interested in knowing the relation between two movies based on user rating.

correlation_matrix = ratings_pivot.corr(method='pearson')
correlation_matrix.head(10)
```

```python
# getting the users who rated Jumanji-movie

Jumanji_rating = ratings_pivot[2]
Jumanji_rating[Jumanji_rating>=0]
```

```python
# finding similar movies to Jumanji movie using Pearson correlation

similar_to_Jumanji = ratings_pivot.corrwith(Jumanji_rating)
corr_Jumanji = pd.DataFrame(similar_to_Jumanji, columns=['PearsonR'])
corr_Jumanji.dropna(inplace=True)
corr_Jumanji.head()
```

```python
Jumanji_corr_summary = corr_Jumanji
```

```python
# getting top 5 most similar movies

Jumanji_corr_summary.sort_values('PearsonR', ascending=False).head(5)
```

```python
# getting the details for most similar movies

movie_corr_Jumanji = pd.DataFrame([2, 92938, 4736, 127204, 101525, 127202], index=np.arange(6),
                                   columns=['movieId'])
summary = pd.merge(movie_corr_Jumanji, movies, on='movieId')
summary
```
